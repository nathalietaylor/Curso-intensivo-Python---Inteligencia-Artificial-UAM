## Séries
Vamos importar o Pandas e explorar o objeto Séries. Uma série no Pandas é muito semelhante a um array no NumPy (na verdade ele é construído em cima do objeto array do Numpy). O que diferencia o array de uma série é que uma serie pode ter rotulos de eixo, o que significa qu eepode ser indexado por um rótulo, em vez de apenas um local numerico. Ele também não precisa conter dados numericos, ele pode conter qualquer objeto Python arbitrário.

```python
import numpy as np
import pandas as pd
```

### Criando um objeto Serie
Você pode converter uma lista, matriz numpy ou dicionário em uma série

```python
labels = ["x", "y", "z"]
my_list = [10,20,30]
arr = np.array([10,20,30])
d ={'a':10,'b':20,'c':30}
```

Utilizando Listas

```python
pd.Series(data=my_list)
```

```python
pd.Series(data=my_list,index=labels)
```

**utilizando numpy array**

```python
pd.Series(arr)
```

```python
pd.Series(arr,labels)
```

**utilizando dicionarios**

```python
pd.Series(d)
```

### Dados em uma Serie
Uma series do pandas pode conter uma variedade de tipos de dados

```python
pd.Series(data=labels)
```

```python
pd.Series(data=[sum,print,len])
```

### Utilizando um índice
A chave para utilizar uma série é entender o funcionamento do seu índice. O Pandas faz uso desses nomes ou numeros de índice, permitindo pesquisas rápidas de informação

```python
ser1 = pd.Series([1,2,3,4], index=["EUA","Alemanha","Russia", "Japão"])
print(ser1)
```

```python
ser2 = pd.Series([1,2,3,4], index=["EUA","Alemanha","Italia", "Japão"])
print(ser2)
```

```python
ser1["EUA"]
```

operações tamém são feitas baseadas nos índices

```python
ser1 + ser2
```

## DataFrames
DataFrames são a base de trabalho do pandas e são diretamente inspirados na linguagem de programação R. Podemos pensar em um DataFrame como um monte de objetos Series reunidos para compartilhar o mesmo indice

```python
from numpy.random import randn #para chamar o método p/ randn e não precisar setar toda a biblioteca
np.random.seed(101)
```

```python
df = pd.DataFrame(randn(5,4), index='A B C D E'.split(),columns='W X Y Z'.split())
print(df)
```

#### Multi-índice e hierarquia de índice

```python
#niveis de indices
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
df= pd.DataFrame(randn(6,2),index=hier_index,columns=['A','B'])
df
```

Como indexar isso? Para hierarquia de índice utilizamos df.loc[ ], se o alvo estivess no eixo das colunas, você utilizaria apenas a notação de colchetes normal df[ ]. Chamar um nível de índice retorna o sub-dataframe.

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```

```python
df.index.names
```

```python
df.index.names = ['Group','Num']
```

```python
df
```

```python
df.xs(('G1',1)) #loc não permite buscar pelo subitem, por exemplo 1, por isso o uso do .xs
```

```python
df.xs(1,level='Num')
```

### Dados ausentes
Vamos mostrar alguns metodos convenientes para lidar com dados ausentes no pandas

```python
df = pd.DataFrame({'A':[1,2,np.nan],'B':[5, np.nan,np.nan], 'C':[1,2,3]}) #{ } sinaliza dicionário
```

```python
df
```

```python
df.dropna() #para dropar NaN em eixo x
```

```python
df.dropna(axis=1) #para coluna
```

```python
df.dropna(thresh=2) # funciona como uma tolerância, nesse caso será dropado quando houver 2 NaN
```

```python
df.fillna(value='fill value') #para preencher valores NaN, é comum por exemplo preencher om a média para não perededr muitos dados, principalmente quando a amostra é pequena
```

```python
df['A'].fillna(value=df['A'].mean()) # subtituir o NaN pela média da coluna A
```
